```python
import matplotlib.pyplot as plt
import numpy as np 
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams.update({'font.size': 12})
```

```python
x = 3 # True slope
a = np.arange(-2,2,0.25)
a = a.reshape(-1,1)
b = a*x + np.random.rand(*a.shape) + np.random.randn(*a.shape)*1.2 #adding noise 
```

```python
plt.plot(a,a*x,Color='k',LineWidth=2,label='True line')
plt.plot(a,b,'x', Color='r', label='Noisy data')

#lets apply SVD decompositon
U,S,VT = np.linalg.svd(a,full_matrices=False) # Economy SVD (U till m)
S = np.diag(S)
#xtilde = VS-1UTb
xtilde = VT.T @ np.linalg.inv(S) @ U.T @ b
plt.plot(a,xtilde*a,'--', Color='b', label='SVD data')

plt.title("Orthogonal proyection with SVD")
plt.xlabel('a')
plt.ylabel('b')
plt.grid(linestyle="--")
plt.legend()
plt.show()
```
