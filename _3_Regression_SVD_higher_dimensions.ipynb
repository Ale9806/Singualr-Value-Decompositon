```python
import matplotlib.pyplot as plt
import numpy as np 
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams.update({'font.size': 12})
from scipy.io import loadmat
```

We are given a data set of heat generated over 12 experiments 

```python
A=loadmat(r"C:\Users\Eduardo\Documents\CARRERA\University_of_washington\data_sets\ingredients")
A=A["A"]
b=loadmat(r"C:\Users\Eduardo\Documents\CARRERA\University_of_washington\data_sets\heat")
b=b["b"]
```

```python
#Getting to know data
print("SHAPES: A:",A.shape," b: ", b.shape)
# 13 experiments
# combinations of 4 ingredients 
```

```python
U,S,VT = np.linalg.svd(A,full_matrices=False) # Economy SVD (U till m)
S = np.diag(S)
#xtilde = VS-1UTb
xtilde = VT.T @ np.linalg.inv(S) @ U.T @ b
plt.plot(b,'-o', Color='k', label='Heat data')
plt.plot(A@xtilde,'-o', Color='b', label='SVD data')

plt.title("Orthogonal proyection with SVD")
plt.xlabel('number of experiment')
plt.ylabel('heat genereated')
plt.grid(linestyle="--")
plt.legend()
plt.show()
```

```python
plt.figure(2)
plt.plot(A[:,1],b,'o', Color='k', label='Heat data')
plt.plot(A[:,1],A@xtilde,'*', Color='b', label='SVD data')

plt.title("Orthogonal proyection with SVD")
plt.xlabel('combination of different ingredients (A[:,0]) first column')
plt.ylabel('heat genereated')
plt.grid(linestyle="--")
plt.show()
```
